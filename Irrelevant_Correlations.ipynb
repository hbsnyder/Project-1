```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
rankings_csv = "Merged/rankings.csv"
```

```python
rankings_df = pd.read_csv(rankings_csv)
rankings_df.head()
```

```python
founded_years_df = rankings_df[["Founded","Wins %"]]
founded_years_df = founded_years_df.set_index("Founded")
founded_years_df.head()
```

```python
wins_founded_df = founded_years_df.sort_values(by='Founded')
wins_founded_df.head()
```

```python
wins_founded_df.plot(kind="bar", figsize=(20,3))
plt.title("Wins % on Year Founded")
plt.show()
plt.tight_layout()
plt.savefig("Images/wins_arena.png")
```

```python
arena_wins_df = rankings_df[["Arena Capacity","Wins %"]]
arena_wins_df = arena_wins_df.set_index("Arena Capacity")
arena_wins_df.head()
```

```python
wins_arena_df = arena_wins_df.sort_values(by='Arena Capacity')
wins_arena_df.head()
```

```python
wins_arena_df.plot(kind="bar", figsize=(20,3))
plt.title("Wins % on Arena Capacity")
plt.show()
plt.savefig("Images/wins_founded.png")
```
